```python
# Step1: Read Dataset
import pandas as pd
df = pd.read_csv('training_set.csv')
df.head()
```

```python
#Step 3: Separate X and Y (Sales Price) values
```

```python
X = df.drop(columns=['SalePrice','Id'])
Y = df[['SalePrice']]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Step 4: perform feature selection
# Cat - Ordinal Encode (Label Encoded)
```

```python
# Separate Categorical(cat) and Continuous(con) features for X
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con 
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pie1 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
##cat_pie1 = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                            ##('ordinal',OrdinalEncoder())]) (not used because NA Values has meaning i.e. Not Alley ,No Basement)
cat_pie1 = Pipeline(steps=[('impute',SimpleImputer(strategy='constant',fill_value='Not_Avialable')),
                            ('ordinal',OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num',num_pie1,con),
                          ('cat',cat_pie1,cat)]).set_output(transform='pandas')
```

```python
X_pre = pre1.fit_transform(X)
```

```python
X_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr,n_features_to_select='auto',direction='backward')
sel.fit_transform(X_pre,Y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)
```

```python
imp_cols
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

```python
# Step 5: Create Pipeline with One Hot Encoder 
```

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pie2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
cat_pie2 = Pipeline(steps=[('impute',SimpleImputer(strategy='constant',fill_value='Not_Avail')),
                           ('ohe',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num',num_pie2,con_sel),
                          ('cat',cat_pie2,cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

```python
# Step 6 : Train Test Split
# 20% Test (Unseen to model)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_sel_pre,Y,test_size=0.2,random_state=21)
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
# Step : 7 Build Model
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
#Step 8 Evaluate the model
model.score(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
def evaluate_model (model,x,y):
    # predict values 
    ypred = model.predict(x)
    from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
    from sklearn.metrics import mean_absolute_percentage_error
    mse = mean_squared_error(y,ypred)
    rmse = mse ** (1/2)
    mae = mean_absolute_error(y,ypred)
    r2 = r2_score(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    # print the result
    print(f'MSE : {mse : .2f}')
    print(f'RMSE : {rmse : .2f}')
    print(f'MAE : {mae : .2f}')
    print(f'MAPE : {mape : .4f}')
    print(f'R2: {r2 :.4f}')      
```

```python
evaluate_model(model,xtrain,ytrain)
```

```python
evaluate_model(model,xtest,ytest)
```
