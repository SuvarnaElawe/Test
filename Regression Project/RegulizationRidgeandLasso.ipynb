```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
#Read from dataset
import pandas as pd
df = pd.read_csv('training_set.csv',na_values=['','NA'],keep_default_na=False)
df.head()
```

```python
# Perform some basic quality check
df.info()
```

```python
df.duplicated().sum()
```

```python
# Step 3: Separate X and Y(SalePrice)
X = df.drop(columns=['Id','SalePrice'])
Y = df[['SalePrice']]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Step 4 : Apply preprocessing on X
#Separate cat and con
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                          ('ohe',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
pre = ColumnTransformer([('num',num_pipe,con),('cat',cat_pipe,cat)]).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
## step apply train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,test_size=0.2,random_state=60)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
# Step 6 : Build Model
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(xtrain,ytrain)
```

```python
model1.fit(xtest,ytest)
```

```python
#Ridge
from sklearn.linear_model import Ridge
model2 = Ridge(alpha=1)
model2.fit(xtrain,ytrain)
```

```python
model2.fit(xtest,ytest)
```

```python
from sklearn.linear_model import Lasso
model3 = Lasso(alpha=1)
model3.fit(xtrain,ytrain)
```

```python
model3.fit(xtest,ytest)
```

```python
#Step 7 : Hyperparameter tuninng -Grid Search cv
# alpha - HyperParameter
# What is optimal value for alpha for Ridge and Lasso for given dataset 
 # SalesPrice(Y)is like weight in project
```

```python
import numpy as np
params = {'alpha':np.arange(start=0.1,stop=20,step=0.1)}
# 1 TO 1000 STEP SIZE =1 ( WHEN DON'T KNOW HOW TO CHOOSE THE TALPHA VALUE)
```

```python
params
```

```python
#  Tuning for Ridge
from sklearn.model_selection import GridSearchCV
rr = Ridge()
gscv1 = GridSearchCV(estimator=rr,param_grid=params,cv=5,scoring='neg_mean_squared_error')
gscv1.fit(xtrain,ytrain)
```

```python
gscv1.best_params_
```

```python
gscv1.best_score_
```

```python
best_ridge = gscv1.best_estimator_
best_ridge
```

```python
best_ridge.score(xtrain,ytrain)
```

```python
best_ridge.score(xtest,ytest)
```

```python
# Tuning for Lasso
ls = Lasso()
gscv2 = GridSearchCV(estimator=ls,param_grid=params,cv=5,scoring='neg_mean_squared_error')
gscv2.fit(xtrain,ytrain)
```

```python
gscv2.best_params_
```

```python
gscv2.best_score_
```

```python
best_lasso = gscv2.best_estimator_
best_lasso
```

```python
best_lasso.score(xtrain,ytrain)
```

```python
best_lasso.score(xtest,ytest)
```

```python
# Cross validation for r2 score
from sklearn.model_selection import cross_val_score
scores_lasso = cross_val_score(best_lasso,xtrain,ytrain,cv=5,scoring='r2')## NEW_MEAN_SQUARED_ERROR
scores_lasso
```

```python
scores_ridge = cross_val_score(best_ridge,xtrain,ytrain,cv=5,scoring='r2')
scores_ridge 
```

```python
scores_lasso.mean()
```

```python
scores_ridge.mean()
```
